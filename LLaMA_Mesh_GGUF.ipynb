```python
# !pip install llama-cpp-python

from llama_cpp import Llama
from huggingface_hub import hf_hub_download

model_path = hf_hub_download(
    repo_id="bartowski/LLaMA-Mesh-GGUF",
    filename="LLaMA-Mesh-Q6_K.gguf"  # or any other quant version
)
```

```python
llm = Llama(
    model_path=model_path,
    n_ctx=2048,
    n_gpu_layers=40
)
```

```python
responses = llm.create_chat_completion(
	messages = [{"role": "user", "content": "Generate a 3D mesh in .obj format of a simple apple with stem and leaf"}],
	max_tokens=768,
	temperature=0.7
)

print(responses)
```

```python
responses
```

```python
print(responses["choices"][0]["message"]["content"])
```

```python
# Example: saving mesh text to .obj file
mesh_output = responses["choices"][0]["message"]["content"]  # Get LLM's mesh output

# Optional: clean up any leading/trailing text (depends on model formatting)
def extract_obj_mesh(text):
    lines = text.strip().splitlines()
    mesh_lines = [line for line in lines if line.startswith(("v ", "f ", "o ", "vn ", "vt ", "#"))]
    return "\n".join(mesh_lines)

obj_data = extract_obj_mesh(mesh_output)

# Save to file
with open("generated_model.obj", "w") as f:
    f.write(obj_data)

print("‚úÖ OBJ file saved as 'generated_model.obj'")
```
